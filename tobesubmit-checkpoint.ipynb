```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
import seaborn as sns
sns.set()
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
train=pd.read_csv("loaddata.csv")
```

```python
train.head()
```

```python
def ls(x):
    if x=='Y':
        return 1
    if x=='N':
        return 0
```

```python
train["Loan_Status"]=train["Loan_Status"].apply(ls)
```

```python
train.head()
```

```python
def pa(x):
    if x=='Urban':
        return 1
    if x=='Rural':
        return 0
```

```python
train["Property_Area"]=train["Property_Area"].apply(pa)
```

```python
train.head()
```

```python
train.isnull().sum()
```

```python
train["Property_Area"].fillna(train["Property_Area"].mean(),inplace=True)
```

```python
train["Property_Area"]=train["Property_Area"].astype(np.int64)
```

```python
train["Credit_History"].fillna(train["Credit_History"].mean(),inplace=True)
```

```python
train["Credit_History"]=train["Credit_History"].astype(np.int64)
```

```python
train.dtypes
```

```python
train["Loan_Amount_Term"].fillna(train["Loan_Amount_Term"].mean(),inplace=True)
```

```python
train["LoanAmount"].fillna(train["LoanAmount"].mean(),inplace=True)
```

```python
def se(x):
    if x=='Yes':
        return 1
    if x=='No':
        return 0
```

```python
train["Self_Employed"]=train["Self_Employed"].apply(se)
```

```python
train["Self_Employed"].fillna(train["Self_Employed"].mean(),inplace=True)
```

```python
train["Self_Employed"]=train["Self_Employed"].astype(np.int64)
```

```python
train.head()
```

```python
def dp(x):
    if x=='Graduate':
        return 1
    if x=='Not Graduate':
        return 0
```

```python
train["Education"]=train["Education"].apply(dp)
```

```python
train["Dependents"]=n.fit_transform(train["Dependents"].astype(str))
```
